# NumPy Basics


```python
import numpy as np
```

## Creating 1D Array

```python
arr = np.array([10, 20, 30, 40, 50])
arr
```

## Shape, Dimension and Data Type


```python
print("Shape:", arr.shape)
print("Dimension:", arr.ndim)
print("Data Type:", arr.dtype)
```

## Creating 2D Array

```python
matrix = np.array([
    [10, 20, 30],
    [40, 50, 60]
])
matrix
```

## Shape and Dimensions of 2D Array

```python
print("Shape:", matrix.shape)
print("Dimension:", matrix.ndim)
```

## Reshape Array

```python
matrix.reshape(3, 2)
```

## Statistical Operations


```python
print("Mean:", np.mean(matrix))
print("Sum:", np.sum(matrix))
print("Maximum:", np.max(matrix))
print("Minimum:", np.min(matrix))
```

## Real-World Example: Traffic Data Analysis

```python
vehicles = np.array([120, 95, 180, 76, 210, 155])

print("Vehicle Count:", vehicles)
print("Average:", np.mean(vehicles))
print("Highest:", np.max(vehicles))
print("Lowest:", np.min(vehicles))
print("Total:", np.sum(vehicles))
```

## Vectorization Example


```python
vehicles + 10
```

## Traffic Simulation

```python
vehicles * 2
```
